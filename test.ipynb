```python
%load_ext autoreload
%autoreload 2

import requests
import random
import time
from selenium.webdriver import Firefox
```

```python
def get_movie_links(browser):
    sel = 'td.titleColumn a'
    link_list = browser.find_elements_by_css_selector(sel)
    return link_list
```

```python
browser = Firefox()
url = "https://www.imdb.com/chart/top?ref_=nv_mv_250"
browser.get(url)
```

```python
link_list = get_movie_links(browser)
```

```python
links_urls = []
for i in link_list:
    links_urls.append(i.get_attribute("href"))
```

```python
#get list of movie actor
def get_movie_cast_list(movie_url):
    browser.get(movie_url)
    sel = 'table.cast_list tr a'
    cast_list = browser.find_elements_by_css_selector(sel)
    names = [i.text for i in cast_list]
    return names
```

```python
#create a list of full urls for cast
def get_full_movies_urls(list_of_urls):
    return [x+'fullcredits/' for x in list_of_urls]

```
