```python
%load_ext autoreload
%autoreload 2

import requests
import random
import time
import lxml.html as lh
import pandas as pd
from selenium.webdriver import Firefox
```

```python
def get_movie_links(browser):
    sel = 'td.titleColumn a'
    link_list = browser.find_elements_by_css_selector(sel)
    return link_list
```

```python
browser = Firefox()
url = "https://www.imdb.com/chart/top?ref_=nv_mv_250"
browser.get(url)
```

```python
link_list = get_movie_links(browser)
```

```python
links_urls = []
for i in link_list:
    links_urls.append(i.get_attribute("href"))
```

```python
links_urls
```

```python
#LSM -- Movie Names List

tr_elements = browser.find_element_by_xpath('//div[@class="lister"]/table/tbody')
movies = tr_elements.text
list_movies = movies.split('\n')
list_movies
```

```python
#cleans up the list of movie URLs into just the links to each movie page
def movie_urls(links_urls):
    url_list = []
    for i in links_urls:
        url_split = i.split('?')
        for e in url_split:
            if e.startswith('https'):
                url_list.append(e)
            else:
                pass
    return url_list
movie_url_list = movie_urls(links_urls)
movie_url_list
```

```python
#TESTING CELL
def get_movie_cast_lists(movie_url_list):
    testing = []
    n = 0
    for i in movie_url_list:
        while n < 5:  
            browser.get(i)
            sel = 'table.cast_list tr a'
            cast_list = browser.find_elements_by_css_selector(sel)
            names = [e.text for e in cast_list]
            testing.append(names)
            time.sleep(1)
            n += 1
    return testing

get_movie_cast_lists(movie_url_list[0:4])
```

```python
#TESTING CELL 2
def get_movie_cast_lists(movie_url_list):
    testing = []
    for i in movie_url_list:
        browser.get(i)
        sel = 'table.cast_list tr a'
        cast_list = browser.find_elements_by_css_selector(sel)
        names = [e.text for e in cast_list]
        testing.append(names)
        time.sleep(1)
    return testing

get_movie_cast_lists(movie_url_list[0:4])
```

```python
top_50_actor_lists = get_movie_cast_lists(movie_url_list[0:4])
```

```python
def list_of_actors(full_cast):
    actor_lists = []
    for i in full_cast:
        data_top10 = i[0:29]
        actor_names = []
        n = 4
        actor_names.append(data_top10[1])
        actor_names.append(data_top10[n])
        actor_names.append(data_top10[n+3])
        x = 10
        while x < 29:
            actor_names.append(data_top10[x])
            x += 3
        actor_lists.append(actor_names)
    return actor_lists

list_of_actors(top_50_actor_lists)
```

```python
top_50_movie_actors = list_of_actors(top_50_actor_lists)
```
